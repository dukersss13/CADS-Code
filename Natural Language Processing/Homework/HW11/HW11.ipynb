```python
import gensim.downloader as api
wv = api.load('word2vec-google-news-300')
```

```python
# Problem 2
words1 = ['beautiful', 'gorgeous', 'dazzling', 'splendid', 'magnificent', 'ugly']
import numpy as np
import pandas as pd

def computeSimilarity(w):
  sM = np.zeros((len(w), len(w)))

  for i in range(len(w)):
    for j in range(len(w)):
      sM[i,j] = wv.similarity(w[i], w[j])
  return sM
```

```python
simMatrix = computeSimilarity(words1)
pd.DataFrame(simMatrix, columns=words1, index=words1)
```

```python
# Problem 3
words2 = ['False', 'fake', 'fraudulent', 'counterfeit', 'spurious', 'true']
simMatrix2 = computeSimilarity(words2)
pd.DataFrame(simMatrix2, columns=words2, index=words2).sum(axis=0)

# True has the smallest distance so it does not belong in this group of words
```

```python
# Problem 4
import tensorflow as tf
import tensorflow_hub as hub

embed = hub.load("https://tfhub.dev/google/Wiki-words-250-with-normalization/2")
```

```python
words = ['Coke', 'Pepsi']
embedding = embed(words)
similarity = np.zeros((len(words), len(words)))

for i in range(len(words)):
  for j in range(len(words)):
    similarity[i,j] = np.inner(embedding[i], embedding[j])

pd.DataFrame(similarity, columns=words, index=words)
```

```python
# Problem 5
from keras.layers import Dense
from keras import Sequential
from keras.optimizers import SGD, Adam
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.model_selection import train_test_split

path2 = "/content/drive/MyDrive/Colab/Admissions.csv"
data = pd.read_csv(path2)
```

```python
X = data.drop(columns=['admit'])
y = data['admit'].values

ss = StandardScaler()
X_scaled = ss.fit_transform(X)
```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(X_scaled, y, test_size=0.3, random_state=24)

enc = OneHotEncoder()
ytrain_ohe = enc.fit_transform(ytrain.reshape(-1,1)).toarray()
ytest_ohe = enc.fit_transform(ytest.reshape(-1,1)).toarray()
```

```python
model = Sequential()
model.add(Dense(units=Xtrain.shape[1]))
model.add(Dense(units=12, activation='relu'))
model.add(Dense(units=2, activation='softmax'))
model.compile(loss='categorical_crossentropy', 
              optimizer=Adam(lr=0.1),
              metrics=['accuracy'])
```

```python
history = model.fit(Xtrain, ytrain_ohe,
                    epochs=50,
                    validation_data=(Xtest, ytest_ohe))
```

```python
# You can see the accuracy is ~70%
y_prob = model.predict(Xtest)
y_pred = y_prob.argmax(axis=1)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
print("Confusion Matrix:\n", confusion_matrix(ytest, y_pred))
```
