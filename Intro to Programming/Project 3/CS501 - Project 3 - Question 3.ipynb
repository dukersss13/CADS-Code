```python
%matplotlib notebook
import matplotlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from datascience import *
matplotlib.rcParams['figure.figsize'] = [8, 6]
```

```python
table1 = Table.read_table('IRIS.csv')
table1
```

```python
x = table1['PetalLength']
y = table1['PetalWidth']
z = table1['Name']
df = pd.DataFrame ({'PetalLength':x,'PetalWidth':y, 'Name':z})
names = df['Name'].astype('category')
name_codes = names.cat.codes
```

```python
%matplotlib notebook
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(x, y, Picker=True, c=name_codes)

def onclick(event):
    Name = df.iloc[event.ind[0]]['Name']
    plt.gca().set_title('Selected item s name is {} '.format(Name))

fig.canvas.mpl_connect('pick_event', onclick)
fig.show()
```
