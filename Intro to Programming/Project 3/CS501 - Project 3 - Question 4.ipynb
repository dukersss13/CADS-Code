```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
from matplotlib import pyplot
```

```python
mean1 = 50
sigma1 = 10  
data1 = np.random.normal(mean1, sigma1, 1000)

mean2 = 75
sigma2 = 50
data2 = np.random.normal(mean2, sigma2, 1000)
```

```python
plt.hist(([data1, data2]), histtype='barstacked', density=True)
graph = np.concatenate((data1,data2))
sn.kdeplot(graph)
```

```python
sn.jointplot(data1, data2, kind='kde', space=0)
```

```python
sn.jointplot(data1, data2, kind='hex', space=0)
```
